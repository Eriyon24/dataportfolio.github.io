````python
import numpy as np

df["loser"] = np.where(df["winner"] == df["r_name"], df["b_name"], df["r_name"])
df
````

````python
#Create winner rows
w = df[["date","winner","loser","method"]].copy()
w["fighter"] = w["winner"]
w["opponent"] = w["loser"]
w["result"] = "win"
````

````python
#Create loser rows
l = df[df["loser"].notna()][["date","winner","loser","method"]].copy()
l["fighter"] = l["loser"]
l["opponent"] = l["winner"]
l["result"] = "loss"
````

````python
import numpy as np
import pandas as pd

# Concatenate w and l to create fighter_df
fighter_df = pd.concat([w, l], ignore_index=True)

# Sort fighter_df by fighter and date
fighter_df = fighter_df.sort_values(["fighter", "date"]).reset_index(drop=True)

# Ensure 'date' column is in datetime format for calculations
fighter_df["date"] = pd.to_datetime(fighter_df["date"], errors="coerce")

# Calculate rest days
fighter_df["days_rest"] = fighter_df.groupby("fighter")["date"].diff().dt.days

# Define bins and labels for rest buckets
bins = [-1, 90, 180, 365, 99999]
labels = ["<90 days", "90–180", "180–365", "365+"]

# Create rest buckets
fighter_df["rest_bucket"] = pd.cut(
    fighter_df["days_rest"],
    bins=bins,
    labels=labels
)

# Display the head of the DataFrame with relevant columns
fighter_df[["fighter", "date", "days_rest", "rest_bucket"]].head(15)
````

````python
#Real rest data
analysis_df = fighter_df.dropna(subset=["rest_bucket"])

analysis_df.head()
````

````python
#Win rates
win_rates = (
    analysis_df.groupby("rest_bucket")["result"]
    .apply(lambda x: (x == "W").mean())
    .reset_index()
)

win_rates["win_rate_pct"] = (win_rates["result"] * 100).round(1)

win_rates
````

````python
# How many fights are there
bucket_counts = (
    analysis_df.groupby("rest_bucket")["result"]
    .count()
    .reset_index()
)
bucket_counts
````

````python
import numpy as np
import pandas as pd

base = df.copy()
base.columns
````

````python
base["loser"] = np.where(
    base["winner"] == base["r_name"], base["b_name"],
    np.where(base["winner"] == base["b_name"], base["r_name"], np.nan)
)
````

````python
cols = ["date","r_name","b_name","winner","loser","method",
        "r_sig_str_landed","b_sig_str_landed"]

w = base[cols].copy()
w["fighter"] = w["winner"]
w["opponent"] = w["loser"]
w["result"] = "W"

l = base[base["loser"].notna()][cols].copy()
l["fighter"] = l["loser"]
l["opponent"] = l["winner"]
l["result"] = "L"

fighter_df = pd.concat([w, l], ignore_index=True)
fighter_df["date"] = pd.to_datetime(fighter_df["date"], errors="coerce")
fighter_df = fighter_df.sort_values(["fighter","date"]).reset_index(drop=True)
````

````python
# Make sure names match cleanly
for c in ["fighter", "r_name", "b_name"]:
    fighter_df[c] = fighter_df[c].astype(str).str.strip()

# Create corner column (NO np.nan)
fighter_df["corner"] = np.select(
    [
        fighter_df["fighter"] == fighter_df["r_name"],
        fighter_df["fighter"] == fighter_df["b_name"]
    ],
    ["R", "B"],
    default=""
)

fighter_df["corner"].value_counts(dropna=False)
````

````python
fighter_df["absorbed_sig"] = np.where(
    fighter_df["corner"] == "R", fighter_df["b_sig_str_landed"],
    np.where(fighter_df["corner"] == "B", fighter_df["r_sig_str_landed"], None)
)

fighter_df[["fighter","corner","absorbed_sig"]].head(10)
````

````python
# Make sure days_rest exists
fighter_df["date"] = pd.to_datetime(fighter_df["date"], errors="coerce")

fighter_df["days_rest"] = fighter_df.groupby("fighter")["date"].diff().dt.days

# Create rest buckets again
bins = [-1, 90, 180, 365, 99999]
labels = ["<90 days", "90–180", "180–365", "365+"]

fighter_df["rest_bucket"] = pd.cut(
    fighter_df["days_rest"],
    bins=bins,
    labels=labels
)

fighter_df[["fighter","date","days_rest","rest_bucket"]].head()
````

````python
analysis_df = fighter_df.dropna(subset=["rest_bucket"])

analysis_df.head()
````

````python
# Win Rate by rest
analysis_df = fighter_df.dropna(subset=["rest_bucket"])

win_rate = (
    analysis_df
    .groupby("rest_bucket")["result"]
    .apply(lambda x: (x == "W").mean())
    .reset_index(name="win_rate")
)

win_rate["win_rate_pct"] = (win_rate["win_rate"] * 100).round(1)

win_rate
````

````python
# Average Damage by rest
damage_by_rest = (
    analysis_df
    .groupby("rest_bucket")["absorbed_sig"]
    .mean()
    .reset_index(name="avg_damage")
)

damage_by_rest
````

````python
#Turning win/loss to numbers for correlation between win/loss
analysis_df["win_num"] = (analysis_df["result"] == "W").astype(int)

analysis_df[["result","win_num"]].head()
````

````python
#Correlation: Damage vs Winning
corr = analysis_df["absorbed_sig"].corr(analysis_df["win_num"])
corr
````

````python
#Correlation with rest too
analysis_df["days_rest"].corr(analysis_df["absorbed_sig"]),
analysis_df["days_rest"].corr(analysis_df["win_num"])
````

````python
#Win rate by rest bucket
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
plt.bar(win_rate["rest_bucket"], win_rate["win_rate_pct"])
plt.title("Win Rate by Rest Bucket")
plt.xlabel("Rest Bucket")
plt.ylabel("Win Rate (%)")
plt.show()
````

````python
#Damage by rest
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
plt.bar(damage_by_rest["rest_bucket"], damage_by_rest["avg_damage"])

plt.title("Average Damage by Rest Bucket")
plt.xlabel("Rest Bucket")
plt.ylabel("Avg Significant Strikes Absorbed (Previous Fight)")
plt.show()
````

````python
#Correlation
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(
    analysis_df["absorbed_sig"],
    analysis_df["win_num"],
    alpha=0.25
)

plt.title("Damage Taken vs Fight Outcome")
plt.xlabel("Significant Strikes Absorbed")
plt.ylabel("Win (1) / Loss (0)")
plt.show()
````
